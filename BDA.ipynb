```r
# Load the necessary libraries
library(ggplot2)

# Load the CSV file
data <- read.csv("/content/drive/My Drive/BDA project 2024/WHR_2023.csv")

# Explore the structure of the data
head(data)
str(data)
summary(data)
```

```r
# Calculate the mean of healthy_life_expectancy
mean_life_expectancy <- mean(data$healthy_life_expectancy, na.rm = TRUE)

# Fill missing values with the mean
data$healthy_life_expectancy[is.na(data$healthy_life_expectancy)] <- mean_life_expectancy

# Check for missing values again
sum(is.na(data))
```

```r
# Basic statistics for happiness_score
mean_happiness <- mean(data$happiness_score)
median_happiness <- median(data$happiness_score)
range_happiness <- range(data$happiness_score)

# Basic statistics for gdp_per_capita
mean_gdp <- mean(data$gdp_per_capita)
median_gdp <- median(data$gdp_per_capita)
range_gdp <- range(data$gdp_per_capita)

# Basic statistics for social_support
mean_social_support <- mean(data$social_support)
median_social_support <- median(data$social_support)
range_social_support <- range(data$social_support)

# Print the results
cat("Happiness Score - Mean:", mean_happiness, " Median:", median_happiness, " Range:", range_happiness, "\n")
cat("GDP per Capita - Mean:", mean_gdp, " Median:", median_gdp, " Range:", range_gdp, "\n")
cat("Social Support - Mean:", mean_social_support, " Median:", median_social_support, " Range:", range_social_support, "\n")
```
